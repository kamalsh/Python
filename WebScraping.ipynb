```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlopen
from bs4 import BeautifulSoup
from pyspark import *
```

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
#url = "http://php.net"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml') # we use lxml when we work with html,css javascript combination
#type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
text=soup.get_text()
print(text)
```

```python
soup.find_all('a')
```

```python
all_links= soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
# Print the first 10 rows for sanity check 
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
    print(row_td)
# type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "html5lib").get_text()
print(cleantext)
```

```python
import re
list_rows =[]
for rows in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('.</?\w*>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
#print(clean2)
print(list_rows)
#type(clean2)
```

```python
df= pd.DataFrame(list_rows)
df.head(20)
```

```python
# Data Manipulation and Cleaning
df1 = df[0].str.split(',',expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1[0] = df1[0].str.strip(']')

df1[1] = df1[1].str.strip(']')
df1[6] = df1[6].str.strip(']')
df1.head(20)
```

```python
col_labels = soup.find_all('th')
col_labels
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2=pd.DataFrame(all_header)
df2
```

```python
df3=df2[0].str.split(',', expand=True)
df3
```

```python
frames = [df3, df1]
df4 = pd.concat(frames)
df4.head(7)
```

```python
df5 = df4.rename(columns = df4.iloc[0])
df5.head()
```

```python
df5.info()
df5.shape
```

```python
df6 = df5.dropna(axis=0)
```

```python
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
df7.rename(columns={'[Place': 'Place'},inplace=True)
df7.rename(columns={' Team]': 'Team'},inplace=True)
# df7.drop([0])
```
